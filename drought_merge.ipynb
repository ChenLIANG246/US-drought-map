```python
from pathlib import Path
import pandas
import geopandas
```

```python
folder = Path("/users/daxiong/desktop/vess projects/drought/DROUGHT data")
gdf=geopandas.GeoDataFrame()
for i in range(2015,2021):
    shapefiles = folder.glob(str(i)+"/USDM_*_M/USDM_*.shp")
    week=1
    for shp in shapefiles:
        df = (geopandas.read_file(shp))
        df['Year']=i
        df['Week']=week
        week+=1
        gdf = pandas.concat([gdf,df],ignore_index=True, sort=False)
gdf.tail()
```

```python
# Save the file
folder_compiled = Path("/users/daxiong/desktop/vess projects/drought/DROUGHT data/compiled")
gdf.to_file(folder_compiled / 'compiled.shp')
gdf.head(10)
```
